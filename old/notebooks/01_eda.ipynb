# 01 — Exploratory Data Analysis: USGS NWIS Groundwater Sites

**Purpose**: Profile and visualize all downloaded NWIS groundwater monitoring sites before any QC or modeling.

Sections:
1. Load & concatenate all state site files
2. Data quality flags (datum, depth, coordinates)
3. Spatial coverage map
4. Site density by state
5. Well depth distribution
6. Land surface elevation distribution
7. Aquifer type and datum breakdown
8. Temporal coverage (construction / inventory dates)
9. Summary statistics table

> **Note**: Ground-level measurements (`data/raw/nwis/*_gwlevels.parquet`) may not yet be available for all states — the `make data` run downloads site metadata and GW levels state-by-state. Re-run `make data` to fill in any missing states. Level data are explored in Section 10 if available.


```python
from __future__ import annotations

import pathlib
import warnings

import geopandas as gpd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np
import pandas as pd
from matplotlib.colors import LogNorm
from matplotlib.gridspec import GridSpec

warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=UserWarning)

# ---- Paths ----
RAW_DIR = pathlib.Path('../data/raw/nwis')
FIG_DIR = pathlib.Path('../figures/eda')
FIG_DIR.mkdir(parents=True, exist_ok=True)

# ---- Scope: all CONUS states ----
CONUS_STATES = [
    'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DC', 'DE', 'FL', 'GA',
    'IA', 'ID', 'IL', 'IN', 'KS', 'KY', 'LA', 'MA', 'MD', 'ME',
    'MI', 'MN', 'MO', 'MS', 'MT', 'NC', 'ND', 'NE', 'NH', 'NJ',
    'NM', 'NV', 'NY', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD',
    'TN', 'TX', 'UT', 'VA', 'VT', 'WA', 'WI', 'WV', 'WY',
]
SCOPE_LABEL = 'CONUS (48 states + DC)'
# Bounding box for maps — EPSG:5070 metres (NAD83 CONUS Albers)
CONUS_XLIM = (-2_400_000,  2_300_000)
CONUS_YLIM = (   270_000,  3_100_000)

# ---- Style ----
plt.rcParams.update({
    'figure.dpi': 120,
    'axes.spines.top': False,
    'axes.spines.right': False,
    'font.size': 14,
    'axes.titlesize': 15,
    'axes.labelsize': 14,
    'xtick.labelsize': 13,
    'ytick.labelsize': 13,
    'legend.fontsize': 13,
})
CMAP_DENSITY = 'YlOrRd'
CMAP_DEPTH   = 'viridis_r'
CMAP_ELEV    = 'terrain'
```

## 1. Load all site files

```python
site_files = sorted(RAW_DIR.glob('*_sites.parquet'))
downloaded_states = sorted(f.stem.replace('_sites', '') for f in site_files)
missing_states = sorted(set(CONUS_STATES) - set(downloaded_states))

print(f'Found {len(site_files)} state site files: {downloaded_states}')
if missing_states:
    print(f'WARNING: {len(missing_states)} states still missing — run `make data` to download:')
    print(f'  {missing_states}')

if not site_files:
    raise RuntimeError("No site parquet files found. Run `make data` first.")

dfs = []
for f in site_files:
    df = pd.read_parquet(f)
    df['state'] = f.stem.replace('_sites', '')
    dfs.append(df)

sites = pd.concat(dfs, ignore_index=True)
print(f'\nTotal sites across all downloaded states: {len(sites):,}')
print(f'States loaded: {sorted(sites["state"].unique())}')
```

```python
# ---- Unit conversions (feet → meters) ----
FT2M = 0.3048
sites['well_depth_m']  = sites['well_depth_va']  * FT2M
sites['alt_va_m']      = sites['alt_va']          * FT2M

# ---- Drop obviously bogus alt_va values (NWIS sentinel = -9999 ft) ----
sites.loc[sites['alt_va'] < -999, 'alt_va_m'] = np.nan

# ---- QC flag columns ----
sites['flag_ngvd29']       = sites['alt_datum_cd'] == 'NGVD29'
sites['flag_deep_well']    = sites['well_depth_m'] > (150)       # >150 m likely confined
sites['flag_no_alt']       = sites['alt_va_m'].isna()
sites['flag_no_coord']     = sites['dec_lat_va'].isna() | sites['dec_long_va'].isna()
sites['flag_any']          = (sites['flag_ngvd29'] |
                               sites['flag_deep_well'] |
                               sites['flag_no_alt'] |
                               sites['flag_no_coord'])

# ---- Sites that would pass QC ----
sites_ok = sites[~sites['flag_any'] & (sites['alt_datum_cd'] == 'NAVD88')].copy()

flag_summary = pd.DataFrame({
    'Flag': ['NGVD29 datum', 'Deep well (>150 m)', 'No elevation (alt_va)', 'No coordinates', 'Any flag'],
    'Count': [
        sites['flag_ngvd29'].sum(),
        sites['flag_deep_well'].sum(),
        sites['flag_no_alt'].sum(),
        sites['flag_no_coord'].sum(),
        sites['flag_any'].sum(),
    ]
})
flag_summary['Pct'] = (flag_summary['Count'] / len(sites) * 100).round(1)
print(flag_summary.to_string(index=False))
print(f'\nSites passing all QC flags (NAVD88, shallow, has elev+coord): {len(sites_ok):,} '
      f'({len(sites_ok)/len(sites)*100:.1f}%)')
```

## 2. QC Flag Summary

```python
fig, ax = plt.subplots(figsize=(7, 3.5))
labels = flag_summary['Flag'].tolist()
counts = flag_summary['Count'].tolist()
colors = ['#e74c3c', '#e67e22', '#f1c40f', '#3498db', '#2c3e50']
bars = ax.barh(labels, counts, color=colors, edgecolor='white', linewidth=0.5)
for bar, pct in zip(bars, flag_summary['Pct']):
    ax.text(bar.get_width() + len(sites)*0.005, bar.get_y() + bar.get_height()/2,
            f'{pct:.1f}%', va='center', fontsize=13, color='#444')
ax.set_xlabel('Number of sites')
ax.set_title('QC flag counts across all downloaded NWIS sites', fontweight='bold')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x/1e3:.0f}k'))
ax.axvline(len(sites), color='gray', linestyle='--', lw=0.8, label=f'Total = {len(sites):,}')
ax.legend(fontsize=13)
plt.tight_layout()
fig.savefig(FIG_DIR / 'fig01_qc_flags.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 3. Spatial Coverage Maps

```python
# Convert to GeoDataFrame for easy spatial plotting
sites_valid_coord = sites.dropna(subset=['dec_lat_va', 'dec_long_va'])
gdf = gpd.GeoDataFrame(
    sites_valid_coord,
    geometry=gpd.points_from_xy(sites_valid_coord['dec_long_va'], sites_valid_coord['dec_lat_va']),
    crs='EPSG:4326'
)
gdf_5070 = gdf.to_crs('EPSG:5070')   # NAD83 CONUS Albers for analysis

# QC-passing sites only
sites_ok_coord = sites_ok.dropna(subset=['dec_lat_va', 'dec_long_va'])
gdf_ok = gpd.GeoDataFrame(
    sites_ok_coord,
    geometry=gpd.points_from_xy(sites_ok_coord['dec_long_va'], sites_ok_coord['dec_lat_va']),
    crs='EPSG:4326'
).to_crs('EPSG:5070')

print(f'Sites with valid coordinates: {len(gdf):,}')
print(f'QC-passing sites with coordinates: {len(gdf_ok):,}')
```

```python
# ---- Load US state boundaries via cartopy naturalearth (geopandas 1.0+) ----
import cartopy.io.shapereader as shpreader
states_shp = shpreader.natural_earth(resolution='110m', category='cultural', name='admin_1_states_provinces')
states_gdf = gpd.read_file(states_shp)
conus = states_gdf[states_gdf['iso_a2'] == 'US'].to_crs('EPSG:5070')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

for ax, (gdf_plot, title, color) in zip(axes, [
    (gdf_5070, f'All downloaded sites\n(n={len(gdf_5070):,})', '#2b7bb9'),
    (gdf_ok,   f'QC-passing sites (NAVD88, shallow, georeferenced)\n(n={len(gdf_ok):,})', '#27ae60'),
]):
    conus.plot(ax=ax, color='#f0f0eb', edgecolor='#cccccc', linewidth=0.4)
    ax.scatter(
        gdf_plot.geometry.x, gdf_plot.geometry.y,
        s=0.5, alpha=0.4, color=color, linewidths=0, rasterized=True
    )
    ax.set_xlim(CONUS_XLIM)
    ax.set_ylim(CONUS_YLIM)
    ax.set_title(title, fontsize=15, fontweight='bold')
    ax.set_axis_off()

plt.suptitle(f'USGS NWIS Groundwater Monitoring Sites — {SCOPE_LABEL}',
             fontsize=16, fontweight='bold', y=1.01)
plt.tight_layout()
fig.savefig(FIG_DIR / 'fig02_spatial_coverage.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ---- Hexbin density map of QC-passing sites ----
fig, ax = plt.subplots(figsize=(14, 6))
conus.plot(ax=ax, color='#efefea', edgecolor='#bbb', linewidth=0.4, zorder=1)

x = gdf_ok.geometry.x.values
y = gdf_ok.geometry.y.values
hb = ax.hexbin(x, y, gridsize=80, cmap=CMAP_DENSITY,
               norm=LogNorm(vmin=1, vmax=None),
               mincnt=1, linewidths=0, zorder=3)

cb = fig.colorbar(hb, ax=ax, shrink=0.6, pad=0.01)
cb.set_label('Sites per hex cell (log scale)', fontsize=13)
ax.set_xlim(CONUS_XLIM)
ax.set_ylim(CONUS_YLIM)
ax.set_title(f'Site density — QC-passing NWIS GW wells (NAVD88, ≤150 m) — {SCOPE_LABEL}',
             fontsize=14, fontweight='bold')
ax.set_axis_off()
plt.tight_layout()
fig.savefig(FIG_DIR / 'fig03_site_density_hexbin.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 4. Site Count by State

```python
state_counts = (
    sites.groupby('state')
    .agg(
        total=('site_no', 'count'),
        navd88=('flag_ngvd29', lambda x: (~x).sum()),
        deep=('flag_deep_well', 'sum'),
    )
    .assign(pct_navd88=lambda d: d['navd88'] / d['total'] * 100)
    .sort_values('total', ascending=True)
)

fig, axes = plt.subplots(1, 2, figsize=(15, max(6, len(state_counts) * 0.25 + 1)))

# Left: total sites per state
ax = axes[0]
bars = ax.barh(state_counts.index, state_counts['total'],
               color='#2b7bb9', edgecolor='white', linewidth=0.3)
ax.set_xlabel('Number of sites')
ax.set_title('Total sites per state', fontweight='bold')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x/1e3:.0f}k'))

# Right: % NAVD88 per state
ax = axes[1]
colors_pct = ['#27ae60' if p >= 50 else '#e74c3c' for p in state_counts['pct_navd88']]
ax.barh(state_counts.index, state_counts['pct_navd88'],
        color=colors_pct, edgecolor='white', linewidth=0.3)
ax.axvline(50, color='#555', linestyle='--', lw=1)
ax.set_xlabel('% sites with NAVD88 datum')
ax.set_title('Fraction of sites with NAVD88 elevation datum\n(green ≥ 50%, red < 50%)', fontweight='bold')

plt.suptitle('NWIS Groundwater Sites by State', fontsize=16, fontweight='bold', x=0.5, y=1.01)
plt.tight_layout()
fig.savefig(FIG_DIR / 'fig04_sites_by_state.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 5. Well Depth Distribution

```python
depth_valid = sites['well_depth_m'].dropna()
depth_valid = depth_valid[(depth_valid > 0) & (depth_valid < 5000)]  # drop clearly bogus values

fig, axes = plt.subplots(1, 3, figsize=(16, 4))

# Linear scale
ax = axes[0]
ax.hist(depth_valid, bins=100, color='#2b7bb9', edgecolor='white', linewidth=0.2)
ax.axvline(150, color='#e74c3c', linestyle='--', lw=1.5, label='150 m (confined flag)')
ax.set_xlabel('Well depth (m)')
ax.set_ylabel('Number of sites')
ax.set_title('All depths (linear scale)')
ax.legend(fontsize=13)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x/1e3:.0f}k'))

# Log scale
ax = axes[1]
ax.hist(depth_valid, bins=np.logspace(np.log10(0.1), np.log10(5000), 80),
        color='#8e44ad', edgecolor='white', linewidth=0.2)
ax.set_xscale('log')
ax.axvline(150, color='#e74c3c', linestyle='--', lw=1.5, label='150 m')
ax.set_xlabel('Well depth (m, log scale)')
ax.set_title('Log-scale depth distribution')
ax.legend(fontsize=13)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x/1e3:.0f}k'))

# CDF
ax = axes[2]
sorted_d = np.sort(depth_valid)
cdf = np.arange(1, len(sorted_d)+1) / len(sorted_d)
ax.plot(sorted_d, cdf, color='#2c3e50', lw=1.5)
ax.axvline(150, color='#e74c3c', linestyle='--', lw=1.5, label='150 m')
p_shallow = (depth_valid <= 150).mean() * 100
ax.axhline(p_shallow/100, color='#e74c3c', linestyle=':', lw=1)
ax.text(155, p_shallow/100 + 0.02, f'{p_shallow:.1f}% ≤ 150 m', fontsize=13, color='#e74c3c')
ax.set_xlabel('Well depth (m)')
ax.set_ylabel('Cumulative fraction')
ax.set_title('CDF of well depth')
ax.legend(fontsize=13)
ax.set_xlim(0, 800)

plt.suptitle(f'Well Depth Distribution  (n={len(depth_valid):,} sites with depth record)',
             fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
fig.savefig(FIG_DIR / 'fig05_well_depth.png', dpi=150, bbox_inches='tight')
plt.show()

print(f'\nWell depth summary (meters):')
print(depth_valid.describe().rename(index=lambda x: x.capitalize()).to_frame('depth_m').T.to_string())
print(f'Sites with depth > 150 m (likely confined): {(depth_valid>150).sum():,} ({(depth_valid>150).mean()*100:.1f}%)')
```

## 10. Temporal Coverage of GW Level Records

Raw time series quality — computed directly from measurement dates, with **no gap filling**.
A site is flagged **sparse** if observed months / possible months < 10 %, or if its longest gap exceeds 36 months.
Only unflagged sites should enter spatial interpolation or model training.


```python

# ---- Load available GW level records (no gap filling) ----
gw_files = sorted(RAW_DIR.glob("*_gwlevels.parquet"))
if not gw_files:
    print("No *_gwlevels.parquet files found — run `make data` to download.")
else:
    print(f"Found {len(gw_files)} state(s) with GW level records: "
          f"{[f.stem.replace('_gwlevels','') for f in gw_files]}")

    dfs_gw = [pd.read_parquet(f) for f in gw_files]
    df_gw_raw = pd.concat(dfs_gw, ignore_index=True)

    # Parse dates and values
    df_gw_raw["date"] = pd.to_datetime(df_gw_raw["lev_dt"], errors="coerce")
    df_gw_raw["dtw_ft"] = pd.to_numeric(df_gw_raw["lev_va"], errors="coerce")
    df_gw_raw = df_gw_raw.dropna(subset=["date", "dtw_ft"]).copy()
    df_gw_raw["year"]  = df_gw_raw["date"].dt.year
    df_gw_raw["month"] = df_gw_raw["date"].dt.month

    # Monthly observation count per site (raw, no imputation)
    monthly_counts = (
        df_gw_raw.groupby(["site_no", "year", "month"])
        .size()
        .reset_index(name="n_obs")
    )

    # Per-site temporal statistics
    def temporal_stats(grp: pd.DataFrame) -> pd.Series:
        ym = (grp["year"] * 12 + grp["month"]).sort_values().values
        n_possible = int(ym[-1] - ym[0] + 1)
        n_obs = len(ym)
        coverage = n_obs / n_possible if n_possible > 0 else 0.0
        gaps = np.diff(ym) - 1
        max_gap = int(gaps.max()) if len(gaps) > 0 else 0
        return pd.Series({
            "n_obs_months": n_obs,
            "n_possible_months": n_possible,
            "coverage_fraction": coverage,
            "max_gap_months": max_gap,
            "record_span_years": n_possible / 12,
            "first_year": int(grp["year"].min()),
            "last_year": int(grp["year"].max()),
        })

    site_cov = monthly_counts.groupby("site_no").apply(
        temporal_stats, include_groups=False
    ).reset_index()

    # Sparsity flags — same thresholds as MIN_COVERAGE_FRACTION / MAX_GAP_MONTHS_FLAG in qc_nwis.py
    MIN_COV = 0.10   # < 10 % of possible months observed → sparse
    MAX_GAP = 36     # > 36 consecutive missing months → long gap

    site_cov["is_sparse"]    = site_cov["coverage_fraction"] < MIN_COV
    site_cov["has_long_gap"] = site_cov["max_gap_months"] > MAX_GAP
    site_cov["usable"]       = ~site_cov["is_sparse"] & ~site_cov["has_long_gap"]

    n_total  = len(site_cov)
    n_sparse = int(site_cov["is_sparse"].sum())
    n_gap    = int(site_cov["has_long_gap"].sum())
    n_usable = int(site_cov["usable"].sum())

    print(f"\nTemporal coverage summary "
          f"({', '.join(f.stem.replace('_gwlevels','') for f in gw_files)}):")
    print(f"  Total sites with GW level records : {n_total:,}")
    print(f"  Sparse (coverage < {MIN_COV:.0%})          : "
          f"{n_sparse:,}  ({n_sparse/n_total*100:.1f}%)")
    print(f"  Long gap (> {MAX_GAP} months)             : "
          f"{n_gap:,}  ({n_gap/n_total*100:.1f}%)")
    print(f"  Usable (neither flag)              : "
          f"{n_usable:,}  ({n_usable/n_total*100:.1f}%)")
    print(f"\n  Median coverage fraction : {site_cov['coverage_fraction'].median():.1%}")
    print(f"  Median record span       : {site_cov['record_span_years'].median():.1f} years")
    print(f"  Median max gap           : {site_cov['max_gap_months'].median():.0f} months")

    # ---- Figure: three coverage diagnostics ----
    fig, axes = plt.subplots(1, 3, figsize=(16, 4))

    ax = axes[0]
    ax.hist(site_cov["coverage_fraction"] * 100, bins=40,
            color='#3498db', edgecolor='white', linewidth=0.2)
    ax.axvline(MIN_COV * 100, color='#e74c3c', linestyle='--', lw=1.5,
               label=f'Sparse threshold ({MIN_COV:.0%})')
    ax.set_xlabel('Coverage fraction (%)')
    ax.set_ylabel('Number of sites')
    ax.set_title('Fraction of possible months\nobserved (no gap filling)')
    ax.legend(fontsize=12)

    ax = axes[1]
    clip_gap = site_cov["max_gap_months"].clip(upper=200)
    ax.hist(clip_gap, bins=np.arange(0, 205, 5),
            color='#e67e22', edgecolor='white', linewidth=0.2)
    ax.axvline(MAX_GAP, color='#e74c3c', linestyle='--', lw=1.5,
               label=f'Long-gap threshold ({MAX_GAP} mo)')
    ax.set_xlabel('Max gap length (months, clipped at 200)')
    ax.set_title('Longest consecutive gap\nbetween observations')
    ax.legend(fontsize=12)

    ax = axes[2]
    ax.hist(site_cov["record_span_years"], bins=40,
            color='#2ecc71', edgecolor='white', linewidth=0.2)
    ax.set_xlabel('Record span (years)')
    ax.set_title('Calendar span first → last\nobservation')

    plt.suptitle(
        f'Temporal Coverage — {len(gw_files)} state(s), '
        f'n={n_total:,} sites  |  no interpolation',
        fontsize=16, fontweight='bold', y=1.02
    )
    plt.tight_layout()
    fig.savefig(FIG_DIR / 'fig10_temporal_coverage.png', dpi=150, bbox_inches='tight')
    plt.show()

    # ---- Figure: usability breakdown pie ----
    fig2, ax2 = plt.subplots(figsize=(5, 5))
    only_sparse = site_cov["is_sparse"]  & ~site_cov["has_long_gap"]
    only_gap    = ~site_cov["is_sparse"] &  site_cov["has_long_gap"]
    both_flags  =  site_cov["is_sparse"] &  site_cov["has_long_gap"]
    flag_counts = [n_usable, int(only_sparse.sum()),
                   int(only_gap.sum()), int(both_flags.sum())]
    flag_labels = ['Usable\n(no flag)', 'Sparse only', 'Long gap only', 'Both flags']
    flag_colors = ['#27ae60', '#e74c3c', '#f39c12', '#8e44ad']
    wedges2, texts2, autotexts2 = ax2.pie(
        flag_counts, labels=flag_labels, colors=flag_colors,
        autopct='%1.1f%%', startangle=90, pctdistance=0.75,
        wedgeprops=dict(edgecolor='white', linewidth=1.5)
    )
    for t in autotexts2:
        t.set_fontsize(13)
    ax2.set_title('Site usability breakdown\n(after temporal QC flags)', fontweight='bold')
    fig2.savefig(FIG_DIR / 'fig11_temporal_flag_breakdown.png', dpi=150, bbox_inches='tight')
    plt.show()
```
